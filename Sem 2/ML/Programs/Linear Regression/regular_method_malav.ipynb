```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([4,6,10,12])
y=np.array([3.0,5.5,6.5,9.0])
n = len(x)
x_mean=np.mean(x)
y_mean=np.mean(y)
#m = slope
numerator = np.sum((x-x_mean)*(y-y_mean))
denominator = np.sum((x-x_mean)**2)
m = numerator/denominator
# intercept=b is 
b= y_mean-m*x_mean
y_pred = m * x + b
print(f"predicted:{y_pred}")
error = y-y_pred
plt.scatter(x, y, color='blue', label="Actual Data")
plt.plot(x, y_pred, color='red', label="Regular regression")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear Regression using regular method")
plt.legend()
plt.grid
plt.show()
```
