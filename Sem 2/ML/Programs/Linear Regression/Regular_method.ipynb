```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def regular_method(x,y):
    n=len(x)
    x_mean=np.mean(x)
    y_mean=np.mean(y)
    # b1 (slope)
    numerator=np.sum((x-x_mean)*(y-y_mean))
    denominator=np.sum((x-x_mean)**2)
    b1=numerator/denominator

    # b0 (intercept)
    b0=y_mean-b1*x_mean

    print(f"Mean of x:{x_mean}")
    print(f"Mean of y:{y_mean}")
    print(f"b0 (intercept):{b0}")
    print(f"b1 (slope):{b1}")

    y_pred=b0 +b1 *x
    print(f"y_pred : {y_pred}")

    error=y-y_pred
    print(f"Error:{error}")

    plt.scatter(x,y,color="blue",label="Actual data")
    plt.plot(x,y_pred,color="red",label="Regression line")

    plt.xlabel('xi')
    plt.ylabel("yi")
    plt.title("LInear regression (Regular method)")
    plt.legend()
    plt.grid(True)
    plt.show()
        
    

```
