```python
import matplotlib.pyplot as plt
```

```python
xi=[4,6,10,12]
yi=[3.0,5.5,6.5,9.0]
m=4
```

```python
xi2=[]
for i in xi:
    xi2.append(i**2)
```

```python
xiyi=[]
for i in range(len(xi)):
    xiyi.append(xi[i]*yi[i])
```

```python
sum_xi=sum(xi)
sum_yi=sum(yi)
sum_xi2=sum(xi2)
sum_xiyi=sum(xiyi)
```

```python
# Least square cofficients
```

```python
a0=(-(sum_xi*sum_xiyi)+(sum_yi*sum_xi2))/((m*sum_xi2)-(sum_xi)**2)
a1=((m*sum_xiyi)-(sum_xi*sum_yi))/((m*sum_xi2)-(sum_xi)**2)
```

```python
print(a0)
print(a1)
```

```python
# Predicatd values
y_pred=[]
for i in xi:
    value=a0+a1*i
    y_pred.append(value)
y_pred
```

```python
# Error
error=[]
for x in range(len(yi)):
    error.append(yi[x] - y_pred[x])
print(f"Error:{error}")
```

```python
# Sum of squared Errors (SSE):
sse=sum([x**2 for x in error])
sse
```

```python
# Plotting
```

```python
plt.scatter(xi,yi,color='red',label="original value")
plt.plot(xi,y_pred,color="blue",label="Fitted Line")
plt.xlabel('xi')
plt.ylabel("yi")
plt.title("LInear regression fit")
plt.legend()
plt.grid(True)
plt.show()

```
