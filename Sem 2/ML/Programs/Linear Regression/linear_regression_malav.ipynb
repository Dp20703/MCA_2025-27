```python
import matplotlib.pyplot as plt
x=[4,6,10,12]
y=[3.0,5.5,6.5,9.0]
n = len(x)
sum_x = sum(x)
sum_y = sum(y)
xiyi=[]
for i in range(n):
    xiyi.append(x[i]*y[i])
sum_xiyi = sum(xiyi)
xi2 = []
for i in range (n):
    xi2.append(x[i]**2)
sum_xi2 = sum(xi2)
m = (n * sum_xiyi - sum_x * sum_y) / (n * sum_xi2 - sum_x ** 2)
print(f"slope is :{m}")
b = (sum_y - m * sum_x) / n
# Predicted y values
y_pred = [m * xi + b for xi in x]
print(f"predicted:{y_pred}")
# error
error=[]
for i in range(n):
    error.append(y[i]-y_pred[i])
    print(f"error:{error[i]}")
#sum squared error 
sse = sum([x**2 for x in error])
print(f"sse is :{sse}")
# Plotting
plt.scatter(x, y, color='blue', label="Actual Points")
plt.plot(x, y_pred, color='red', label="Best Fit Line")

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear Regression using Least Squares")
plt.legend()

plt.show()
```
